Q11. Are head circumference data for babies of smoking mothers normally distributed?
Q12. What is the significance value for the above on the Shapiro-Wilk test?

```python
import matplotlib.pyplot as plt
from scipy.stats import shapiro

head_circumference_data = smoking_data['Headcirc']

# Plot histogram
plt.hist(head_circumference_data, bins='auto', color='#0504aa', alpha=0.7, rwidth=0.85)
plt.title('Head Circumference Distribution for Babies of Smoking Mothers')
plt.xlabel('Head Circumference')
plt.ylabel('Frequency')
plt.show()

stat, p_value = shapiro(head_circumference_data)
print("Shapiro-Wilk Test p-value:", p_value)

if p_value > 0.05:
    print("The head circumference data is likely normally distributed.")
else:
    print("The head circumference data is not likely normally distributed.")
```

Q17. Based on the dataset you have, how confident can you be in saying that a baby's birth weight will be +/- 1 standard deviation from the mean?

```python
birth_weight_data = df['Birthweight']

mean_birth_weight = birth_weight_data.mean()
std_dev_birth_weight = birth_weight_data.std()

within_1_std_dev = ((birth_weight_data >= (mean_birth_weight - std_dev_birth_weight)) &
                    (birth_weight_data <= (mean_birth_weight + std_dev_birth_weight))).mean() * 100

print("Mean Birth Weight:", mean_birth_weight)
print("Standard Deviation of Birth Weight:", std_dev_birth_weight)
print(f"Percentage of Birth Weights within ±1 Standard Deviation: {within_1_std_dev:.2f}%")
```

To assess how confident we can be in saying that a baby's birth weight will be within ±1 standard deviation from the mean, we used the empirical rule (68-95-99.7 rule) for normally distributed data. This rule states that for a normal distribution:

Approximately 68% of the data falls within ±1 standard deviation from the mean.
Approximately 95% of the data falls within ±2 standard deviations from the mean.
Approximately 99.7% of the data falls within ±3 standard deviations from the mean.


Q18. Based on the dataset you have, what is the probability that the birth weight for a baby of a smoking mother will be less than 4.2 kg?

```python
from scipy.stats import norm

birth_weight_smoking = df[df['smoker'] == 1]['Birthweight']

mean_birth_weight_smoking = birth_weight_smoking.mean()
std_dev_birth_weight_smoking = birth_weight_smoking.std()

probability_less_than_4_2_kg = norm.cdf(4.2, loc=mean_birth_weight_smoking, scale=std_dev_birth_weight_smoking) * 100

print(f"Probability that the birth weight for a baby of a smoking mother is less than 4.2 kg: {probability_less_than_4_2_kg:.2f}%")
```

Q19. Are data for length of baby of non-smoking mothers normally distributed?

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, probplot


length_non_smoking = df[df['smoker'] == 0]['Length']

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(length_non_smoking, bins=20, kde=True, color='blue')
plt.title('Histogram of Baby Length for Non-Smoking Mothers')

plt.subplot(1, 2, 2)
probplot(length_non_smoking, plot=plt)
plt.title('Q-Q Plot for Baby Length of Non-Smoking Mothers')

plt.tight_layout()
plt.show()

statistic, p_value = shapiro(length_non_smoking)

alpha = 0.05
if p_value > alpha:
    print("The length data for babies of non-smoking mothers appears to be normally distributed.")
else:
    print("The length data for babies of non-smoking mothers does not appear to be normally distributed.")
```

Q21. What is the standard score for the length of a baby of 48.5cm for non-smoking mothers?

```python
length_non_smoking = df[df['smoker'] == 0]['Length']

mean_length_non_smoking = length_non_smoking.mean()
std_dev_length_non_smoking = length_non_smoking.std()

baby_length = 48.5
z_score = (baby_length - mean_length_non_smoking) / std_dev_length_non_smoking

print(f"Standard Score (Z-score) for a baby with length 48.5 cm for non-smoking mothers: {z_score:.2f}")
```
