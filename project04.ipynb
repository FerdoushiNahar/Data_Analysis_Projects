4.	Are there any significant outliers?

```python
import numpy as np
import pandas as pd

# Read your dataset into a pandas DataFrame
data = pd.read_csv('/content/drive/MyDrive/kreativ/PickUpLines.csv')  # Replace 'your_dataset.csv' with the actual filename

# Extract the column containing the variable you want to check for outliers
variable_column = data['Receptivity']

# Calculate the z-score for each data point
z_scores = np.abs((variable_column - variable_column.mean()) / variable_column.std())

# Define a threshold for identifying outliers (e.g., z-score > 3 is considered an outlier)
threshold = 3

# Identify outliers
outliers = data[z_scores > threshold]

if len(outliers) > 0:
    print("Significant outliers found:")
    print(outliers)
else:
    print("No significant outliers found.")
```


The code provided checks for outliers using the z-score method, which identifies data points that deviate significantly from the mean of the dataset in terms of standard deviations. If the z-score of a data point exceeds a certain threshold (commonly set to 3), it is considered an outlier.

In this case, the code determined that no significant outliers were found based on the specified threshold. This means that none of the data points in the 'Receptivity' column deviate significantly from the mean of the column in terms of standard deviations.

5.	How is your dependent variable distributed in each cell?

```python
import pandas as pd
from scipy.stats import chi2_contingency


# Define the dependent variable
dependent_var = "Receptivity"

# Define the independent variables
independent_vars = ["PickUp", "Scent", "Gender", "Age", "Ethnicity", "YearInSchool", "Contraceptive"]

# Perform chi-square test for each independent variable
for var in independent_vars:
    # Create a contingency table
    contingency_table = pd.crosstab(index=data[dependent_var], columns=data[var])

    # Perform the chi-square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Output the p-value and variable name
    print(f"Variable: {var}, P-value: {p}")

    # Interpret the result
    alpha = 0.05
    if p < alpha:
        print("The scores in each cell are dependent (reject the null hypothesis)")
    else:
        print("The scores in each cell are independent (fail to reject the null hypothesis)")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency


# Contingency table (cross-tabulation)
contingency_table = pd.crosstab(data['PickUp'], data['Receptivity'])

# Perform chi-square test for independence
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies table:")
print(expected)
```

Chi-square statistic: 64.685
p-value: 0.0659
The p-value is greater than the typical significance level of 0.05. Therefore, we fail to reject the null hypothesis that there is no relationship between "PickUp" and "Receptivity."

```python
from scipy.stats import levene

# Extract Receptivity data for each level of PickUp
receptivity_pickup1 = data[data["PickUp"] == 1]["Receptivity"]
receptivity_pickup2 = data[data["PickUp"] == 2]["Receptivity"]

# Perform Levene's test for homogeneity of variance
statistic, p_value = levene(receptivity_pickup1, receptivity_pickup2)

# Output the result
if p_value < 0.05:
    print("There is evidence of heterogeneity of variance.")
else:
    print("There is no evidence of heterogeneity of variance.")
```

7.	Is there homogeneity or heterogeneity of variance?

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import levene


# Rename PickUp levels
data["PickUp"] = data["PickUp"].map({1: "Cute Direct", 2: "Direct Direct"})

# Extract Receptivity data for each level of PickUp
receptivity_pickup1 = data[data["PickUp"] == "Cute Direct"]["Receptivity"]
receptivity_pickup2 = data[data["PickUp"] == "Direct Direct"]["Receptivity"]

# Perform Levene's test for homogeneity of variance
statistic, p_value = levene(receptivity_pickup1, receptivity_pickup2)

# Plot the distribution of Receptivity for each level of PickUp
plt.figure(figsize=(10, 6))
plt.hist(receptivity_pickup1, alpha=0.5, label="Cute Direct", color="blue", edgecolor="black")
plt.hist(receptivity_pickup2, alpha=0.5, label="Direct Direct", color="orange", edgecolor="black")
plt.axvline(receptivity_pickup1.mean(), color='blue', linestyle='dashed', linewidth=1, label='Mean Cute Direct')
plt.axvline(receptivity_pickup2.mean(), color='orange', linestyle='dashed', linewidth=1, label='Mean Direct Direct')
plt.xlabel("Receptivity")
plt.ylabel("Frequency")
plt.title("Distribution of Receptivity by PickUp")
plt.legend()
plt.grid(True)
plt.show()

# Output Levene's test result
if p_value < 0.05:
    print("There is evidence of heterogeneity of variance (p-value < 0.05).")
else:
    print("There is no evidence of heterogeneity of variance (p-value >= 0.05).")
```
